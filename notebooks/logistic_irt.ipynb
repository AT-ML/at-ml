## In this notebook, we demostrate how to use some collected ML experiment results to train a Logitic IRT model, and visualise the item characteristic curves.

```python
import sys; sys.path.insert(0, '..')
import sys; sys.path.insert(0, '../atml')
import os
import numpy
import pandas
import matplotlib.pyplot
import joblib
from atml.irt import Logistic_IRT
from atml.measure import BS
from atml.visualisation import get_logistic_curve
%matplotlib inline
```

## Load some previous ML results as Pandas dataframe.

```python
res = pandas.read_csv('./res_base.csv', index_col=0)
res
```

## Get all the possible index values for both datasets and models, and specify the corresponding evalution measure.

```python
dataset_list = res['data_idx'].unique()
model_list = res['model_idx'].unique()
measure = BS()
```

## Specify the Logitic IRT model and fit the model to the experiment results.

```python
mdl = Logistic_IRT()
```

```python
mdl.fit(dataset_list, 
        res['data_idx'].to_numpy().astype('int'), 
        model_list, 
        res['model_idx'].to_numpy().astype('int'), 
        BS.transform(res[measure.name].to_numpy()))
```

## Save the model for adaptive testing

```python
joblib.dump(mdl, 'logistic_mdl.gz')
```

## Draw the item characteristic curve for each dataset

```python
data_dict = {0: 'iris',
             1: 'digits',
             2: 'wine'}

for i in range(0, 3):
    get_logistic_curve(i,  data_dict[i], mdl, res, measure)
```
