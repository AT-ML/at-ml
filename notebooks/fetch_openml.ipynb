## In this notebook, we will demostrate how to collect exsisting machine learning experiments and performance measurements from OpenML. The collected results can also be used to train the base IRT models for the adpative testing process.

```python
import sys; sys.path.insert(0, '..')
import openml
import pandas
import atml.exp
```

## For OpenML, we need to specify two parameters to specify the model-dataset pair. 

## The first parameter is the dictionary of OpenML data_id and task_id:
https://www.openml.org/search?type=data 

https://www.openml.org/search?type=task 

## The second parameter is the dictionary of OpenML flow_id:
https://www.openml.org/search?type=flow

```python
openml_dict = {0: ('adult', 1590, 7592),
               1: ('breast-cancer', 13, 13)}

flow_dict = {1172: 0,
             1726: 1}
```

## A evaluation measure can be specified as a string, and all the available measures can be seen at:

https://www.openml.org/search?type=measure

```python
measure = 'predictive_accuracy'
```

## The results  can be collected with the get_openml_testing function.

```python
res = atml.exp.get_openml_testing(openml_dict, flow_dict, measure)
```

## We can check the results with Pandas dataframe.

```python
res
```

## Save the results (to be used later for IRT training)

```python
res.to_csv('./res_openml.csv')
```
